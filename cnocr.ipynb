```
import matplotlib.pyplot as plt
from cnocr import CnOcr

ocr = CnOcr()

pdfImg = 'ocr/pdf.png'
im = plt.imread(pdfImg)
plt.imshow(im)
plt.show()
pdf = ocr.ocr(pdfImg)
print("cnocr Predicted PDF:", pdf)

mpaasImg = 'ocr/mpaas.png'
im = plt.imread(mpaasImg)
plt.imshow(im)
plt.show()
mpaas = ocr.ocr(mpaasImg)
mpaas
print("cnocr Predicted mpaas:", mpaas)

rpImg = 'ocr/rp.png'
im = plt.imread(rpImg)
plt.imshow(im)
plt.show()
rp = ocr.ocr(rpImg)
print("cnocr Predicted rp:", rp)
```

发现对大段文字识别可以，但是对表格文字识别很差。所以可能需要先识别出每部门文字区域，然后分别调用ocr识别。
